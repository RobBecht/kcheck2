```python
import pandas as pd
```

Reading in data from a CSV that I will want to aggregate later. To do this I will need to clean some of the inputs to remove special characters and ensure proper data type. Column Saved was verified with the sources Data dictionary as incorrectly containing negative values.

```python
df = pd.read_csv("Louisville_Metro_KY_-_Fire_Property_Damage.csv")
df.head()
```

My choice of cleaning actions are to: change data type to str, correct negative values to 0 for future aggregation, change data types back to numeric so agg functions work, and rename Saved column to Protected because an extra feature felt like a good choice.

```python
df['Saved'] = df['Saved'].astype(str)

df.loc[df['Saved'].str.startswith('-'), 'Saved'] = '0'

df['Saved'] = pd.to_numeric(df['Saved'])

df.rename(columns={'Saved': 'Protected'}, inplace=True)

df.head()
```
